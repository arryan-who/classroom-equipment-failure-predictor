```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/equipment_data.csv")

df.head()
```

```python
print("Dataset shape:", df.shape)

df.info()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="equipment_type",
    data=df
)

plt.title("Distribution of Equipment Types")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="failure_within_30_days",
    data=df
)

plt.title("Failure Occurrence Distribution")

plt.show()
```

```python
# plot the correlation heatmap of the features

plt.figure(figsize=(8,5))

sns.countplot(
    x="equipment_type",
    hue="failure_within_30_days",
    data=df
)

plt.title("Failure Distribution by Equipment")

plt.show()
```

```python
# plot the boxplot of room temperature by failure occurrence

plt.figure(figsize=(6,4))

sns.boxplot(
    x="failure_within_30_days",
    y="daily_usage_hours",
    data=df
)

plt.title("Daily Usage vs Failure")

plt.show()
```
