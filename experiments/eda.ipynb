```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/equipment_data.csv")

df.head()
```

```python
df.info()
```

```python
# plot the distribution of the target variable
plt.figure(figsize=(6,4))

sns.countplot(
    x="failure_occurred",
    data=df
)

plt.title("Failure Distribution")
plt.show()
```

```python
# plot the distribution of the target variable by equipment type

plt.figure(figsize=(8,5))

sns.countplot(
    x="equipment_type",
    hue="failure_occurred",
    data=df
)

plt.title("Failure by Equipment Type")
plt.show()
```

```python
# plot the correlation heatmap of the features

plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# plot the boxplot of room temperature by failure occurrence

plt.figure(figsize=(6,4))

sns.boxplot(
    x="failure_occurred",
    y="room_temperature",
    data=df
)

plt.title("Temperature vs Failure")
plt.show()
```

```python
# plot the boxplot of daily usage hours by failure occurrence

plt.figure(figsize=(6,4))

sns.boxplot(
    x="failure_occurred",
    y="daily_usage_hours",
    data=df
)

plt.title("Usage vs Failure")
plt.show()
```
